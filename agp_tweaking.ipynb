```python
# Create a new column with Super IDs
hap_counters = {'Hap_1': 1, 'Hap_2': 1}
unloc_counter = 1
current_hap = None
group_ids = []

# Iterate over rows to assign groupID
for _, row in df.iterrows():
    if row["Orientation"] == "proximity_ligation":
        group_ids.append(group_id)
        continue

    paint_value = row["paint"]

    if paint_value in hap_counters:
        # New group for this haplotype
        current_hap = paint_value
        group_id = f"Super_{hap_counters[current_hap]}_{paint_value}"
        hap_counters[current_hap] += 1  # Increment the counter for the haplotype
    elif paint_value == "Unloc" and current_hap:
        # "Unloc" case, no increment in main hap counter
        group_id = f"Super_{hap_counters[current_hap] - 1}_{current_hap}_Unloc_{unloc_counter}"
        unloc_counter += 1  # Increment the Unloc counter within the group
    elif current_hap:
        # Continuation of the current group without new haplotype value
        group_id = f"Super_{hap_counters[current_hap] - 1}_{current_hap}"
        unloc_counter = 1  # Reset Unloc counter if no "Unloc" value

    group_ids.append(group_id)

# Assign the groupID column to the dataframe
df["groupID"] = group_ids
```

```python
import os
import pandas as pd
from Bio import SeqIO

# Create folders "hap_1" and "hap_2" if they don't exist
os.makedirs("hap_1", exist_ok=True)
os.makedirs("hap_2", exist_ok=True)

# Assume you have loaded your DataFrame as df, for example:
# df = pd.read_csv("data.csv", header=None)

# Initialize variables to keep track of the current FASTA file and output file.
current_name = None         # Holds the current FASTA file path (from col6 when a string)
fasta_dict = None           # Holds the records from the current FASTA file
current_output_path = None  # Holds the path to the currently used output file
current_folder = None       # Current folder ("hap_1" or "hap_2")
last_output_filename = None # Tracks the last output filename

# Iterate over every row in the DataFrame.
for idx, row in df.iterrows():
    # Column indices (0-indexed):
    # col6 -> row[5], col7 -> row[6], col8 -> row[7], col9 -> row[8], col15 -> row[14] (if exists)
    col6 = row[5]

    if isinstance(col6, str):
        # Determine folder based on substring in col6:
        if "H1" in col6:
            current_folder = "hap_1"
        elif "H2" in col6:
            current_folder = "hap_2"

        # Load the new FASTA file (if available)
        if os.path.exists(col6):
            try:
                fasta_dict = SeqIO.to_dict(SeqIO.parse(col6, "fasta"))
            except Exception as e:
                print(f"Error parsing FASTA file {col6} on row {idx}: {e}")
                fasta_dict = None
        else:
            print(f"FASTA file {col6} not found. Skipping row {idx}.")
            fasta_dict = None

        # Determine the output filename: if column15 exists and is not null, use it; otherwise, use the basename of col6.
        if pd.notna(row[14]):
            output_filename = str(row[14])
        else:
            output_filename = os.path.basename(col6)

        # Build the full output file path in the appropriate folder.
        new_output_path = os.path.join(current_folder, output_filename)

        # Read strand from column9 (index 8)
        sign = str(row[8]).strip()

        # Get start and end positions from columns 7 and 8 (indexes 6 and 7).
        try:
            start_idx = int(row[6])
            end_idx = int(row[7])
        except Exception as e:
            print(f"Error converting start/end indices on row {idx}: {e}")
            continue

        # Extract the sequence region from the loaded FASTA.
        extracted_seq = ""
        if fasta_dict is not None and len(fasta_dict) > 0:
            # Here we assume the FASTA file contains one record.
            record = list(fasta_dict.values())[0]
            if sign == '+':
                # Adjust for 1-indexed positions.
                extracted_seq = record.seq[start_idx - 1 : end_idx]
            elif sign == '-':
                # For '-', extract from end_idx to start_idx.
                extracted_seq = record.seq[end_idx - 1 : start_idx]
            else:
                print(f"Unrecognized strand '{sign}' on row {idx}. Skipping extraction.")
        else:
            print(f"No FASTA data loaded for row {idx}.")

        # If the output filename has changed, create a new file (with header), else append without header.
        if last_output_filename != output_filename:
            current_output_path = new_output_path
            last_output_filename = output_filename
            with open(current_output_path, "w") as outf:
                # Write header (FASTA headers must start with a '>')
                header = f">Row{idx}_{os.path.basename(col6)}_{start_idx}_{end_idx}\n"
                outf.write(header)
                # Write the extracted sequence as a continuous string (no added newlines)
                seq_str = str(extracted_seq)
                outf.write(seq_str)
        else:
            # Same output file: append the sequence directly (without header or newlines)
            with open(current_output_path, "a") as outf:
                seq_str = str(extracted_seq)
                outf.write(seq_str)

    else:
        # If col6 is numeric, do not change the current folder or output file.
        # Append 100 'N's directly (continuously) to the current output FASTA file, if available.
        if current_output_path is not None:
            with open(current_output_path, "a") as outf:
                outf.write("N" * 100)
        else:
            print(f"No current output file defined for row {idx} with numeric col6.")

```
