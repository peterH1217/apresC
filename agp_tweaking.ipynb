```python
import os
import pandas as pd
from Bio import SeqIO
import gzip

# Create output folders if they don't exist.
os.makedirs("hap_1", exist_ok=True)
os.makedirs("hap_2", exist_ok=True)

# Load the single FASTA file that contains all records.
fasta_file = "/content/mMarFla1.HiC.combined.decontam.20240513.fasta.gz"  # your file name
with gzip.open(fasta_file, "rt") as handle:
    fasta_dict = SeqIO.to_dict(SeqIO.parse(handle, "fasta"))

# Assume your DataFrame is loaded as df, e.g.:
# df = pd.read_csv("data.csv", header=None)

# Fixed output file names for each folder.
output_file_h1 = os.path.join("hap_1", "Hap_1.fasta")
output_file_h2 = os.path.join("hap_2", "Hap_2.fasta")

# We'll track, for each output file, the last header (after cleaning) used,
# and also whether we already inserted an empty line for rows without header.
last_header_dict = {"Hap_1.fasta": None, "Hap_2.fasta": None}
empty_line_inserted = {"Hap_1.fasta": False, "Hap_2.fasta": False}

# Iterate over each row in the DataFrame.
for idx, row in df.iterrows():
    # Columns (0-indexed):
    #   col6: row[5] – record header key in the FASTA file (or numeric)
    #   col7: row[6] – start index
    #   col8: row[7] – end index
    #   col9: row[8] – strand
    #   col15: row[14] – (optional) output header string (e.g., "Super_2_Hap_2", "Super_2_Hap_2_Unloc_1")

    # Decide which output file (and folder) to use.
    # If col15 is available, use its prefix; otherwise, decide based on row[5]'s content.
    if pd.notna(row[14]):
        raw_header = str(row[14])
        if raw_header.startswith("Hap_1_"):
            current_output_path = output_file_h1
            current_key = "Hap_1.fasta"
        elif raw_header.startswith("Hap_2_"):
            current_output_path = output_file_h2
            current_key = "Hap_2.fasta"
        else:
            current_output_path = output_file_h1  # default to hap_1
            current_key = "Hap_1.fasta"
    else:
        if isinstance(row[5], str) and "H2" in row[5]:
            current_output_path = output_file_h2
            current_key = "Hap_2.fasta"
        else:
            current_output_path = output_file_h1
            current_key = "Hap_1.fasta"

    # Process if row[5] is a string (record header key) or numeric.
    if isinstance(row[5], str):
        # Look up the record from the loaded FASTA file using row[5].
        record = fasta_dict.get(row[5])
        if record is None:
            print(f"Record '{row[5]}' not found in {fasta_file} (row {idx}).")
            continue

        # Get start and end indices from columns 7 and 8.
        try:
            start_idx = int(row[6])
            end_idx = int(row[7])
        except Exception as e:
            print(f"Error converting indices on row {idx}: {e}")
            continue

        # Get strand from column 9.
        sign = str(row[8]).strip()
        extracted_seq = ""
        if sign == '+':
            extracted_seq = record.seq[start_idx - 1: end_idx]
        elif sign == '-':
            extracted_seq = record.seq[end_idx - 1: start_idx]
        else:
            print(f"Unrecognized strand '{sign}' on row {idx}.")
        seq_str = str(extracted_seq)

        # Determine the output header, if available.
        if pd.notna(row[14]):
            # Remove the folder prefix from raw_header.
            if current_key == "Hap_1.fasta":
                out_header = raw_header.replace("Hap_1_", "", 1)
            elif current_key == "Hap_2.fasta":
                out_header = raw_header.replace("Hap_2_", "", 1)
            else:
                out_header = raw_header
        else:
            out_header = None  # Indicates no header for this row.

        # Write to the output file.
        if out_header is not None:
            if last_header_dict[current_key] != out_header:
                # New header: start a new FASTA record.
                with open(current_output_path, "a") as outf:
                    # Optionally add an empty line if this is not the very first record.
                    if last_header_dict[current_key] is not None:
                        outf.write("\n")
                    outf.write(f">{out_header}")
                    outf.write(seq_str)
                last_header_dict[current_key] = out_header
                empty_line_inserted[current_key] = False
            else:
                # Same header: append the sequence fragment continuously.
                with open(current_output_path, "a") as outf:
                    outf.write(seq_str)
        else:
            # For rows without a header (col15 missing):
            if not empty_line_inserted[current_key]:
                with open(current_output_path, "a") as outf:
                    outf.write("\n")  # Insert an empty line before starting no-header entries.
                empty_line_inserted[current_key] = True
            with open(current_output_path, "a") as outf:
                outf.write(seq_str)

    else:
        # If row[5] is numeric, simply append 100 "N"s to the current output file.
        if current_output_path is not None:
            with open(current_output_path, "a") as outf:
                outf.write("N" * 100)
        else:
            print(f"No current output file defined for row {idx} with numeric row[5].")

# After processing all rows, print the contents of the output files.
print("Contents of Hap_1.fasta:")
with open(output_file_h1, "r") as f:
    print(f.read())

print("\nContents of Hap_2.fasta:")
with open(output_file_h2, "r") as f:
    print(f.read())
```
